```python
fig = plt.figure(figsize=(8,3.5))
gs = gridspec.GridSpec(1,2,width_ratios=(2,1))

ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])
ax0.set_xlim(-1,8.5)
ax0.set_ylim(1.3,9.5)

x = np.array([0,2,2,4,4,4,4,6,6,6,6,6,6,6,6])
y = np.array([5,7,3,8,6,4,2,8.5,7.5,6.5,5.5,4.5,3.5,2.5,1.5])
ax0.scatter(x,y,facecolor='white',edgecolor='black',zorder=3)
ax0.plot([-1,0],[5,5],c='black',zorder=1)
ax0.plot([0,0.5,2,2.5,4,4.5,6],[5,7,7,8,8,8.5,8.5],c='black',zorder=1)
ax0.plot([0,0.5,2,2.5,4,4.5,6],[5,3,3,2,2,1.5,1.5],c='black',zorder=1)
ax0.plot([2,2.5,4,4.5,6],[7,6,6,5.5,5.5],c='black',zorder=1)
ax0.plot([2,2.5,4,4.5,6],[3,4,4,4.5,4.5],c='black',zorder=1)
ax0.plot([4,4.5,6],[6,6.5,6.5],c='black',zorder=1)
ax0.plot([4,4.5,6],[4,3.5,3.5],c='black',zorder=1)
ax0.plot([4,4.5,6],[8,7.5,7.5],c='black',zorder=1)
ax0.plot([4,4.5,6],[2,2.5,2.5],c='black',zorder=1)
x = np.linspace(6.5,8.0,1000)
f1 = 0.5/(0.3*np.sqrt(2*np.pi))*np.exp(-0.5*((x-7)/0.3)**2)
f2 = 0.5/(0.3*np.sqrt(2*np.pi))*np.exp(-0.5*((x-7.2)/0.3)**2)
f3 = 0.5/(0.2*np.sqrt(2*np.pi))*np.exp(-0.5*((x-7.1)/0.2)**2)
f4 = 0.5/(0.25*np.sqrt(2*np.pi))*np.exp(-0.5*((x-7.15)/0.25)**2)
f5 = np.zeros(len(x))
f6 = np.zeros(len(x))
f7 = np.zeros(len(x))
f8 = np.zeros(len(x))
f5[(x>7.2)&(x<7.6)] = 0.2/(0.4)
f6[(x>6.9)&(x<7.3)] = 0.2/(0.4)
f7[(x>6.8)&(x<7.2)] = 0.2/(0.4)
f8[(x>7.0)&(x<7.4)] = 0.2/(0.4)
ax0.plot(x,f1+8.5)
ax0.plot(x,f2+7.5)
ax0.plot(x,f3+6.5)
ax0.plot(x,f4+5.5)
ax0.plot(x,f5+4.5)
ax0.plot(x,f6+3.5)
ax0.plot(x,f7+2.5)
ax0.plot(x,f8+1.5)
ax0.spines[['right','top','left','bottom']].set_visible(False)
ax0.xaxis.set_ticks([])
ax0.yaxis.set_ticks([])
ax1.plot(x,f1, alpha=0.5)
ax1.plot(x,f2, alpha=0.5)
ax1.plot(x,f3, alpha=0.5)
ax1.plot(x,f4, alpha=0.5)
ax1.plot(x,f5, alpha=0.5)
ax1.plot(x,f6, alpha=0.5)
ax1.plot(x,f7, alpha=0.5)
ax1.plot(x,f8, alpha=0.5)
ax1.plot(x,(0.09*f1+0.21*f2+0.09*f3+0.21*f4+0.06*f5+0.14*f6+0.06*f7+0.14*f8),c='black',label='combined')

ax0.annotate('0.6',xy=(1,7),va='center',backgroundcolor='w')
ax0.annotate('0.4',xy=(1,3),va='center',backgroundcolor='w')
ax0.annotate('0.5',xy=(3,8),va='center',backgroundcolor='w')
ax0.annotate('0.5',xy=(3,6),va='center',backgroundcolor='w')
ax0.annotate('0.5',xy=(3,4),va='center',backgroundcolor='w')
ax0.annotate('0.5',xy=(3,2),va='center',backgroundcolor='w')
ax0.annotate('0.3',xy=(5,8.5),va='center',backgroundcolor='w')
ax0.annotate('0.7',xy=(5,7.5),va='center',backgroundcolor='w')
ax0.annotate('0.3',xy=(5,6.5),va='center',backgroundcolor='w')
ax0.annotate('0.7',xy=(5,5.5),va='center',backgroundcolor='w')
ax0.annotate('0.3',xy=(5,4.5),va='center',backgroundcolor='w')
ax0.annotate('0.7',xy=(5,3.5),va='center',backgroundcolor='w')
ax0.annotate('0.3',xy=(5,2.5),va='center',backgroundcolor='w')
ax0.annotate('0.7',xy=(5,1.5),va='center',backgroundcolor='w')
ax0.annotate('w=0.09',xy=(7.8,8.8))
ax0.annotate('0.21',xy=(7.8,7.8))
ax0.annotate('0.09',xy=(7.8,6.8))
ax0.annotate('0.21',xy=(7.8,5.8))
ax0.annotate('0.06',xy=(7.8,4.8))
ax0.annotate('0.14',xy=(7.8,3.8))
ax0.annotate('0.06',xy=(7.8,2.8))
ax0.annotate('0.14',xy=(7.8,1.8))

ax1.grid(True,alpha=0.5)
ax1.set_ylabel(r'$f(x)$')
ax1.set_xlabel(r'$x$')

ax1.annotate('Combined',xy=(7.21,0.59),xytext=(7.4,0.7),arrowprops=dict(facecolor='black',arrowstyle='->'))
ax1.annotate('Individual',xy=(7.21,0.79),xytext=(7.4,0.9),arrowprops=dict(facecolor='black',arrowstyle='->'))

ax0.text(-0.9, 0.2, '(a)');
ax1.text(6, -.2, '(b)');

plt.tight_layout()
```
